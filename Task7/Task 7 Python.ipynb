### Mean

Mean, along with median and mode, are one of the most common and basic statistics term, which is commonly called the $\textbf{measures of central tendency}$. 
Mean is the average of the data points.
Mathematically, if $x_1, x_2, \dots, x_n$ are the set of data points we have, then the $\textbf{mean}$ of $x_1, \dots, x_n$ is
$$\frac{1}{n} \sum_{i = 1}^{n} x_i$$

In Python, we do this by using the numpy library, i.e. np.mean()

```python
# check mean of income
avg = np.mean(dataset['Mthly_HH_Income'])
print("The average of Monthly HH Income is", avg)
```

### Median

Median is the middle of the set of data points. 

Mathematically, if there are $n$ data points $x_1 \le x_2 \le \dots \le x_n$, then: 
<ol>
    <li> If $n$ is odd, then the median is $x_{\frac{n + 1}{2}}$, i.e. the $\frac{n + 1}{2}$th data after being ordered.
<li> If $n$ is even, then the median is $\frac{x_{\frac{n}{2}} + x_{\frac{n}{2} + 1}}{2}$, i.e. the average of $\frac{n}{2}$th and $\left( \frac{n}{2} + 1 \right)$th data after being ordered.
</ol>

In python, we do this by using the numpy library, i.e. np.median()

```python
# check median of income 
med = np.median(dataset['Mthly_HH_Income'])
print("The median of Monthly HH Income is", med)
```

### Mode

<p>Mode are the data points that appeared most in a series of observation.</p>

Mathematically, suppose that $x_1, \dots, x_n$ are the data points, with $y_1, \dots y_m$ being the unique distinct data from the $x_i$s. Suppose that $y_i$ appear $n_i$ times in the data points, where we have $n_1 + \dots + n_m = n$. Then, if $n_1 \le \dots \le n_m$, then $y_{n_k}$ is a mode whenever $n_k = \max_{1 \le i \le m} n_i$

In Python, we do this by using the statistics library, i.e. st.mode()

```python
# mode of No of Fly Members
import statistics as st
temp = st.mode(dataset['No_of_Fly_Members'])
print("The mode of Number of Fly Members is", temp)
```

### Variance

Variance is a measure of how data points vary from the mean.

Mathematically, if $x_1, \dots, x_n$ are the set of data points with mean $\mu$, then the variance is calculated by 
$$\sigma^2 = \frac{1}{n} \sum_{i = 1}^{n} (x_i - \mu)^2$$

In Python, we do this by using the statistics library, st.variance()

```python
# check variance of income
var = st.variance(dataset['Mthly_HH_Income'])
print("The variance of Monthly HH Income is", var)
```

### Standard Deviation

Standard deviation is a measure of the amount of variation or how wide-spread of a set of data values. 

Mathematically, if $x_1, \dots, x_n$ are the set of data points with mean $\mu$, then the standard deviation is calculated by
$$\sigma = \sqrt{\frac{1}{n} \sum_{i = 1}^{n}(x_i - \mu)^2}$$

In Python, we do this by using the statistics library, st.stdev()

```python
# check standard deviation of income
s = st.stdev(dataset['Mthly_HH_Income'])
print("The variance of Monthly HH Income is", s)
```

### Correlation 

Correlation is a measure of how things are related. We usually do this by looking at the correlation coefficient, which is a number between -1 and 1 that tells you how strongly related the two things are. 
If the correlation coefficient is between $0$ and $1$, then it has positive correlation, as in when one variable changes, the other variables change in the same direction. When the correlation coefficient is $0$, both variables are said to have no correlation at all. If it is between $-1$ and $0$, then it has negative correlation, as in when one variable changes, the other variables change in the opposite direction.

We will give a formula for Pearson's correlation coefficient, which is the most commonly used correlation coefficient as Pearson allows for strong inferences, which is goot to measure linear relationship between variables. 

Let $x_1, \dots, x_n$ and $y_1, \dots, y_n$ be the data points. Then
$$r = \displaystyle \frac{n \sum_{i = 1}^{n} x_i y_i - \left( \sum_{i = 1}^n x_i \right) \left( \sum_{i = 1}^n y_i \right)}{\sqrt{\left( n \sum_{i = 1}^{n} x_i^2 - \left( \sum_{i = 1}^n x_i \right)^2 \right) \left( n \sum_{i = 1}^n y_i^2 - \left( \sum_{i = 1}^n y_i \right)^2 \right)}}$$

We usually do this by using the numpy library, np.corrcoeff()

```python
# calculate the correlation 
temp = np.corrcoef(dataset["Mthly_HH_Income"], dataset["Mthly_HH_Expense"])
print(temp)
```

### Normal Distribution

The normal distribution is a probability distribution widely used as it is commonly used to model phenomena such as physical characteristics. 

A normal distribution with mean $\mu$ and variance $\sigma^2$ is often denoted as $N(\mu, \sigma^2)$, and its probability density function is 
$$\frac{1}{\sigma \sqrt{2 \pi}}e^{-\frac{(x - \mu)^2}{2 \sigma^2}}$$

### Feature of Normal Distribution

Normal Distribution comes with a lot of interesting feature, as follow: 
<ul>
<li> The mean, median and the mode are the same, it is the center.
<li> The expected value of this distribution is also the center.
<li> The frequency curve is bell shaped -- and it has perfect bilateral symmetry
</ul>

### Positively Skewed & Negatively Skewed Normal Distribution and Effect on Mean, Median and Mode due to Skewness

Skewness is the degree of distortion from the symmetrical bell curve or the normal distribution. It measures the lack of symmetry in data distribution.

We often measure skewness using either of the following two methods: 
<ul>
   <li>Pearson mode skewness, is usually used when there's a strong mode which is exhibited by the sample data. The formula is given by
       $$\text{Skewness} = \frac{\mu - \text{Mo}}{\sigma}$$
       where $\mu$ is the average of the data points, $\sigma$ is the standard deviation, and $\text{Mo}$ is the mode. 
   <li>Pearson median skewness, if there are multiple modes (weak modes) in the data points. The formula is given by
       $$\text{Skewness} = \frac{3 \mu - \text{Md}}{\sigma}$$
       where $\mu, \sigma$ is defined as above and $\text{Md}$ is the median of the data points.
</ul>
As shown above, a positive skew diagram means that the skewness value is positive and hence the mean and median of the data points is greater than the mode. In this case, the diagram is slightly skewed towards the left of the diagram, indicating that there are a lot of points below average or median. 

Similarly, a negative skew diagram means that the skewness value is negative and hence the mean and median of the data points is less than the mode. In this case, the diagram is slightly skewed towards the right of the diagram, indicating that there are a lot of points above average or median. 

### QQ Plot 

A QQ plot is a plot of the quantiles of two distributions against each other, or a plot based on estimates of the quantiles. The pattern of points in the plot is used to compare the two distributions.

```python
# create 100 random values that follow a normal distribution
np.random.seed(0)
data = np.random.normal(0,1, 1000)

import statsmodels.api as sm
import matplotlib.pyplot as plt

fig = sm.qqplot(data, line='45')
plt.show()
```

### Box Cox

Box Cox Transformation is a transformation that attempts to transform your data such that the transformed values are normally distributed. The transformation are typically
$$Y(\lambda) = \begin{cases} \frac{Y^{\lambda} - 1}{\lambda} &\text{when } \lambda \not= 0 \\ \log(y) &\text{when } \lambda = 0 \end{cases}$$

```python
from scipy.stats import boxcox 
import seaborn as sns 
data = np.random.exponential(size=1000)
transformed_data, best_lambda = boxcox(data) 

#plot the distribution of the transformed data values
sns.displot(transformed_data, kind = 'kde') ;
```
